```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
data = pd.read_csv("Documents/Movie_reviews.txt",sep = '\t', names=['Reviews','Comments'])   
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data['Reviews'].value_counts()
```

```python
def clean_review(review):
    str=' '.join(word for word in review.split() if word.lower() not in stopwords.words('english'))
    return str
```

```python
data['Comment']=data['Comments'].apply(clean_review)
```

```python
data
```

```python
data['Comment'] = data['Comment'].str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
data['Comment']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=3000)
```

```python
reviews=cv.fit_transform(data['Comment']).toarray()
```

```python
data['Reviews'] = data['Reviews'].replace(['1','0','2'], [1,0,2]).infer_objects(copy=False)
```

```python
data['Reviews'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
reviews_train, reviews_test, sent_train, sent_test = train_test_split(
    reviews, data['Reviews'],test_size=0.2, stratify=data['Reviews'], random_state=42)
```

```python
model.fit(reviews_train,sent_train)
```

```python
predict=model.predict(reviews_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(sent_test, predict, labels=model.classes_)
plt.figure(figsize=(5,5))
display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['Negative','Positive','Neutral'])
display.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(sent_test, predict, target_names=['Negative','Positive','Neutral']))
```

```python
from sklearn.metrics import accuracy_score
predict = model.predict(reviews_test)
accuracy = accuracy_score(sent_test, predict)
print("Model Accuracy: {:.2f}%".format(accuracy * 100))
```
