```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
data=pd.read_csv("reviews.txt",sep = '\t', names =['Reviews','Comments'])
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data['Reviews'].value_counts()
```

```python
def clean_review(review):
    str=' '.join(word for word in review.split() if word.lower() not in stopwords.words('english'))
    return str
```

```python
data['Comment']=data['Comments'].apply(clean_review)
```

```python
data
```

```python
data['Comment'] = data['Comment'].str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
#Wordcloud

from wordcloud import WordCloud
```

```python
#draw Wordcloud

reviews =' '.join(word for word in data['Comment'][data['Reviews'] == 0].astype(str))
wordcloud = WordCloud(height=600,width=1000,max_font_size=100)
plt.figure(figsize= (15,12))
plt.imshow(wordcloud.generate(reviews),interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
reviews =' '.join(word for word in data['Comment'][data['Reviews'] == 1].astype(str))
wordcloud = WordCloud(height=600,width=1000,max_font_size=100)
plt.figure(figsize= (15,12))
plt.imshow(wordcloud.generate(reviews),interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=3000)
```

```python
reviews=cv.fit_transform(data['Comment']).toarray()
```

```python
data['Reviews']=data['Reviews'].replace(['1','0'],[1,0])
```

```python
data['Reviews'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
reviews_train,reviews_test,sent_train,sent_test= train_test_split(reviews,data['Reviews'],test_size=0.2)
```

```python
model.fit(reviews_train,sent_train)
```

```python
predict=model.predict(reviews_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(sent_test, predict,labels=model.classes_)
plt.figure(figsize=(5,5))
display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
display.plot()
plt.show()
```
